```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/nifty50.csv")

df.head()
```

```python
df.info()
```

```python
sector_counts = df["Sector"].value_counts()

sector_counts
```

```python
sector_counts.plot(kind="bar")

plt.title("Sector Distribution of NIFTY 50 Companies")
plt.xlabel("Sector")
plt.ylabel("Number of Companies")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
sector_counts.plot(kind="pie", autopct="%1.1f%%")

plt.title("Sector Composition of NIFTY 50")

plt.ylabel("")

plt.show()
```

```python
plt.savefig("../visuals/sector_pie_chart.png")
```
